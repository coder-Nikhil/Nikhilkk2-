```python
#### threshold = 0.9
def high_corr(data,thresh):
    res = set()
    corr = data.corr()
    for i in range(len(corr.columns)):
        for j in range(i):
            if corr.iloc[i,j]>=thresh:
                res.add(corr.columns[i])
    return list(res)

multicoll_cols=high_corr(df,0.9)
print(multicoll_cols)
print(len(multicoll_cols))
```
